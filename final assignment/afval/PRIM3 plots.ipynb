```python
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open("data\ScenMultiMORDM", 'rb') as file2:
    Scenarios2 = pickle.load(file2)
```

```python
with open("data\ScenariosOpenExplo2.0", 'rb') as file3:
    Scenarios3 = pickle.load(file3)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 15))

sns.barplot(dict(Scenarios3[1]), ax=ax1)
ax1.set_title('Worst-case A.5 (before MORDM)')
ax1.tick_params(labelrotation=90)

sns.barplot(dict(Scenarios2[1]), ax=ax2)
ax2.set_title('Worst-case A.5 (after MORDM)')
ax2.tick_params(labelrotation=90)

sns.barplot(dict(Scenarios3[0]), ax=ax3)
ax3.set_title('Worst-case Total (before MORDM)')
ax3.tick_params(labelrotation=90)

sns.barplot(dict(Scenarios2[0]), ax=ax4)
ax4.set_title('Worst-case Total (after MORDM)')
ax4.tick_params(labelrotation=90)

plt.show()
```
